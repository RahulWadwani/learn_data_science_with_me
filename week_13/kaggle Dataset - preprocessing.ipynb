# removing the null values 

```python
del df['sub_title']
del df['notebooks']
del df['topics']
df.head()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.to_csv('v2.csv',index=False)
```

```python
df.head()
```

```python
#2) Data preprocessing 
```

```python
views = []
for i in df['views']:
    views.append(i.replace(',',''))

df['views'] = views
```

```python
downloads = []
for i in df['downloads']:
    downloads.append(i.replace(',',''))

df['downloads'] = downloads
```

```python
size = []
for i in df['size']:
    if (i[-2:] == 'kB'):
        size.append(round(int(i[:-2])/1024,4))
    elif (i[-2:] == 'MB'):
        size.append(int(i[:-2]))
    elif (i[-2:] == 'GB'):
        size.append(int(i[:-2])*1024)
    else:
        size.append(round(int(i[:-2])/(1024*1024),4))
df['size'] = size
```
