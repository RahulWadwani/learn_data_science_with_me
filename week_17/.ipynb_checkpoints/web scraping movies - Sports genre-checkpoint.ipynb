```python
import requests
import pandas as pd 
import numpy as np 
from bs4 import BeautifulSoup 
link  = 'https://www.imdb.com/search/title/?genres=sport'
res   = requests.get(link)
soup  = BeautifulSoup(res.content,'lxml')
```

```python
for i in soup.find_all('div',class_ = 'lister-item mode-advanced'):
    try:
        movie_id       = (i.find('div',class_ = 'lister-item-content').find('a').get('href').split('/')[-2])
    except:
        movie_id = np.nan
    try:
        movie_name     = (i.find('div',class_ = 'lister-item-content').find('a').text.strip())
    except:
        movie_name = np.nan
    try:
        year           = (i.find('div',class_ = 'lister-item-content').find_all('span')[1].text.strip())
    except:
        year = np.nan
    try:
        certificate    = (i.find('p').find('span',class_ = 'certificate').text.strip())
    except:
        certificate = np.nan
    try:
        duration       = (i.find('p').find('span',class_ = 'runtime').text.strip())
    except:
        duration = np.nan
    try:
        genre          = (i.find('p').find('span',class_ = 'genre').text.strip())
    except:
        genre = np.nan
    try:
        rating         = (i.find('div',class_ = 'ratings-bar').find('strong').text.strip())
    except:
        rating = np.nan
    try:
        description    = (i.find_all('p',class_ = 'text-muted')[1].text.strip())
    except:
        description = np.nan
    try:
        if ('Stars' in str(i.find_all('p')[2])):
            star_id    = ','.join([stars.get('href').split('/')[-2] for stars in BeautifulSoup(str(i.find_all('p')[2]),'lxml').find_all('a')])
            star_name  = ','.join([stars.text for stars in BeautifulSoup(str(i.find_all('p')[2]),'lxml').find_all('a')])
    except:
        star_id = np.nan
        star_name = np.nan
        
    
    print(star_id,star_name)
    
```
