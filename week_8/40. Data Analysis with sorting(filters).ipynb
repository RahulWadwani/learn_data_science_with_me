# 1.2) Top 10 apps based on their reviews


```python
df.sort_values(by = 'review',ascending = False).head(10)
```

```python
apps = set()
for i in df.sort_values(by = 'review',ascending = False).values:
    apps.add(i[0])
    if len(apps) == 10:
        break
list(apps)
```

```python
c= 1
for i in apps:
    print(c,':',i)
    c += 1
```

#  1.3) top 10 apps based on their reviews (sorted format)

```python
lst = []
for app  in apps:
    for i in df.values:
        if i[0]==app:
            lst.append([i[0],i[3]])
            break
df_pr = pd.DataFrame(lst,columns = ['name','reviews'])
c = 1
for i in df_pr.sort_values(by = 'reviews', ascending = False).values:
    print(c," : ",i[0])
    c += 1
```
