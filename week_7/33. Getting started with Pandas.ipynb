# Creating a dataframe

```python
df = pd.read_csv('Data.csv',header = None)

name = df[0]
hindi = df[1]
english = df[2]
maths = df[3]
science = df[4]
ss = df[5]

del df[0]
del df[1]
del df[2]
del df[3]
del df[4]
del df[5]
```

```python
df['name']= name
df['Hindi'] = hindi
df['English'] = english
df['maths'] = maths
df['science'] = science
df['ss'] = ss
```

# filing the null values 

```python
df
```

```python
df['name'] = df['name'].fillna('Anonymous')
df['Hindi'] = df['Hindi'].fillna(72.20)
df['English'] = df['English'].fillna(74.00)
df['maths'] = df['maths'].fillna(70.20)
df['science'] = df['science'].fillna(69.80)
df['ss'] = df['ss'].fillna(75.90)
```

# Removing the null values 

```python
df.dropna()# dropping the null values from the dataframes
```

```python
df.dropna().isnull().sum()
```
