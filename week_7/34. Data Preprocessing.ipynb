```python
import pandas as pd 
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
def typecast(col_name):
    lst = []

    for i in df[col_name]:
        lst.append(int(i))
    df[col_name] = lst
```

```python
typecast('English')
typecast('maths')
typecast('Hindi')
typecast('science')
typecast('ss')
```

```python
df
```

# Adding new data into dataframe

```python
total = []
avg = []

for i in range(len(df)):
    total.append(df.iloc[i,1:].sum())
    avg.append(df.iloc[i,1:].sum()/5)
    
```

```python
df['Total'] = total
df['avg'] = avg
```

```python
df
```

# Sorting students data based on marks 

```python
df_pr = df.sort_values(by = 'Total',ascending = False).head()
df_pr.to_csv('Topper_in_Exam.csv',index = False)
```

# topper in Hindi based on marks 

```python
df_pr = df.sort_values(by = 'Hindi',ascending = False).head()
df_pr.to_csv('Topper_in_Hindi.csv',index = False)
```

# topper in English based on marks 

```python
df_pr = df.sort_values(by = 'English',ascending = False).head()
df_pr.to_csv('Topper_in_English.csv',index = False)
```

# topper in maths based on marks 

```python
df_pr = df.sort_values(by = 'maths',ascending = False).head()
df_pr.to_csv('Topper_in_Maths.csv',index = False)
```

# topper in science based on marks 

```python
df_pr = df.sort_values(by = 'science',ascending = False).head()
df_pr.to_csv('Topper_in_Science.csv',index = False)
```

# topper in ss based on marks 

```python
df_pr = df.sort_values(by = 'ss',ascending = False).head()
df_pr.to_csv('Topper_in_ss.csv',index = False)
```
