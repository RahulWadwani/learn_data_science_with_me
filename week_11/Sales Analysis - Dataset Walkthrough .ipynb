```python
import pandas as pd
import numpy as np 
import datetime
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Data-Science-Specialization/main/Week%20-%2011/Data%20Preprocessing%20-%20Date/sales.csv')
```

```python
df.head()
```

### 1.checking null values 

```python
df.isnull().sum()
```

### 2. Adding total billing amount columns 
#### 2.1. Removing duplicate rows with uncertain  values 

```python
data = []
c = 0
for i in df.values:
    if (i[2] != 'Quantity Ordered'):
        data.append(i)
data = pd.DataFrame(data,columns = ('order_id','product','quantity' ,'price','date','address','month'))
data.head()
```

#### 2.2 creating total billing amount column 

```python
total = []
for i in data.values:
    qn = float(i[2])
    pr = float(i[3])
    total.append(round(qn*pr,2))
data['total'] = total
data.head()
```

### 3. preprocessing date
#### 3.1 adding date and time column

```python
date = []
time = []
for i in data.values:
    date.append(i[4].split(' ')[0])
    time.append(i[4].split(' ')[1])
data['date'] = date
data['time'] = time
data.head()
```

#### 3.2) Adding hour

```python
hour = []
for i in data['time']:
    hour.append(i.split(':')[0])
data['hour'] = hour
data.head()
```

#### 3.3) day month and year column 

```python
date = []
month = []
year = []
for i in data['date']:
    month.append(i.split('/')[0])
    date.append(i.split('/')[1])
    year.append(i.split('/')[2])
data['date_'] = date
data['month_'] = month 
data['year_'] = year
data.head()
    
```

#### 3.4) Adding day column (ex.mon,tues)

```python
day = []
for i in data['date']:
    month = int(i.split('/')[0])
    year = int(i.split('/')[2])
    date = int(i.split('/')[1])
    day.append(datetime.date(year,month,date).strftime('%a'))
data['day'] = day
data.head()
```

### 4. preprocessing the address
#### 4.1) address walkthrough 

```python
data['address']
```

#### 4.2) adding the pincode column

```python
pincode = []
for i in data['address']:
    pincode.append(i.split(' ')[-1])
    
data['pincode'] = pincode
data.head()
    
```

#### 4.3) validate the pincodes 

```python
pincode = []
for i in data['pincode']:
    if len(i) == 5:
        pincode.append(int(i))
        
data['pincode'] = pincode 
data.head()
```

#### 4.4) Adding states columns 

```python
state =[]
for i in data['address']:
    if len(i.split(' ')[-2]) == 2:
        state.append(i.split(' ')[-2])
data['state'] = state
data.head()
    
    
```

#### 4.5) Adding city columns

```python
city =[]
for i in data['address']:
    city.append(i.split(',')[-2].strip())
        
data['city'] = city
data.head()
    
    
```

#### 4.6) Adding update address

```python
street =[]
for i in data['address']:
    street.append(i.split(',')[0].strip())
data['street'] = street
data.head()
```
