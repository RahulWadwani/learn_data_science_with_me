# 2) Based on State

```python
mkdir State
```

```python
for i in df['state'].unique():
    path = ('State/'+i+'.csv')
    df_ = (df[df['state']==i])
    df_.to_csv(path,index = False)
```

# 3) Based on city - based on product solds

```python
data = []
for city in df['city'].unique():
    
    q = 0
    o = 0
    a = 0
    for i in df.values:
        if (i[11] == city):      # i[11] == city 
            o += 1               # count the order
            q += i[2]            # i[2] == quantity , count the quantity
            a += i[4]            # i[4] == total    , count the total
    data.append([city,o,q,int(a)])
df_ = pd.DataFrame(data,columns = ['City','Orders','Quantity','Amount'])
df_.to_csv('City.csv',index= False)
```

# 4) City 

```python
mkdir city
```

```python
for i in df['city'].unique():
    path = 'city/'+i+'.csv'
    df_ = df[df['city'] == i]
    df_.to_csv(path,index = False)
```

# 5) Products

```python
data = []
for i in df['product'].unique():
    df_ = df[df['product']==i]
    data.append([i,
                 df_['price'].mode().values[0],
                 len(df_),df_['quantity'].sum(),
                 df_['quantity'].max(),
                 int(df_['total'].sum())
                ])
df_ = pd.DataFrame(data,columns = ['Name','Price','Orders','Quantity','Max_quantity_per_user','Total_amount'])
```

```python
df_.to_csv('Product.csv',index = False)
```
