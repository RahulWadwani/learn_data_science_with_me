```python
import matplotlib.pyplot as plt
```

# Level - 1

```python
marks = [60,40,60,40]
plt.pie(marks)
plt.show()
```

# Level -2 (Add labels)

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
plt.pie(marks,labels = labels )
plt.show()
```

# level -3 (adding angles)

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
```

```python
plt.pie(marks,labels =labels,startangle = 0)
plt.show()
```

# Level - 4 | Adding Explode

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
explodes = [0,0,0.0,0.2]
plt.pie(marks, labels= labels, explode = explodes)

plt.show()
```

#  Level -5 (adding shadows)

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
explodes = [0,0,0.0,0.2]
plt.pie(marks, labels= labels, explode = explodes,shadow = True)
plt.show()
```

# Level -6 |Adding specific colours 

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
colors = ['#000FFF','#FFF000','#F0F0F0','#0F0F0F']
explodes = [0,0,0.0,0.2]
plt.pie(marks, labels= labels, explode = explodes,shadow = True,colors = colors)
plt.show()
```

# Level 7 | add Legend

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
explodes = [0,0,0.0,0.2]
plt.pie(marks, labels= labels, explode = explodes,shadow= True)
plt.legend()

plt.show()
```

# Level -8 |adding only legend

```python
marks = [45,50,55,60]
plt.pie(marks, explode = explodes,shadow= True)
plt.legend(['Hindi','English','Maths','Science'])

plt.show()
```

# Level - 9 |adding title

```python
marks = [45,50,55,60]
plt.pie(marks, explode = explodes,shadow= True)
plt.legend(['Hindi','English','Maths','Science'],title = 'Subjects')

plt.show()
```

# Level 10 | Adding Numbers

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Maths','Science']
explodes = [0,0,0.0,0.2]
plt.pie(marks, labels= labels, explode = explodes,shadow = True,autopct ='%1.2f%%')
plt.show()
```

# 2. Scatter Plot


# marks distribution for girls  

```python
girls = [88,90,91,89,95]

standard = [6,7,8,9,10]

plt.scatter(standard,girls,edgecolor = 'orange',marker = 's',c = 'pink',linewidth = 2)



plt.legend(['Girls','Boys'])

plt.xlabel('Marks')
plt.ylabel('Class')
plt.title('Marks distribution of girls')
plt.show()
```

# marks distribution for boys 

```python
boys = [78,87,88,83,90]

standard = [6,7,8,9,10]

plt.scatter(standard,boys,edgecolor = 'grey',marker = '^',c = 'green',linewidth = 2)


plt.legend(['Girls','Boys'])

plt.xlabel('Marks')
plt.ylabel('Class')
plt.title('Marks distribution of boys')
plt.show()
```

# level - 1 | Adding Boys and Girls 

```python
girls = [88,90,91,89,95]
boys = [78,87,88,83,90]

standard = [6,7,8,9,10]

plt.scatter(standard,girls,edgecolor = 'orange',marker = 's',c = 'pink',linewidth = 2)
plt.scatter(standard,boys,edgecolor = 'grey',marker = '^',c = 'green',linewidth = 2)


plt.legend(['Girls','Boys'])

plt.xlabel('Marks')
plt.ylabel('Class')
plt.title('Marks distribution of boys and girls in different classes')
plt.show()
```

# Level - 2 | Adding Minimum and maximum

```python
girls = [88,90,91,89,95]
boys = [78,87,88,83,90]

minimum = [0,0,0,0,0]
maximum = [100,100,100,100,100]

standard = [6,7,8,9,10]

plt.scatter(standard,girls,edgecolor = 'orange',marker = 's',c = 'pink',linewidth = 2)
plt.scatter(standard,boys,edgecolor = 'grey',marker = '^',c = 'green',linewidth = 2)
plt.scatter(standard, minimum,c = 'white')
plt.scatter(standard, maximum,c = 'white')
plt.legend(['Girls','Boys'])

plt.xlabel('Marks')
plt.ylabel('Class')
plt.title('Marks distribution of boys and girls in different classes')
plt.show()
```
