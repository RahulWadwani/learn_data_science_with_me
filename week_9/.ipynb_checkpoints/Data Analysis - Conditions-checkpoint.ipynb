```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('playstore.csv')
```

```python
df.head()
```

# 1) all apps comes under art_and_design

```python
df[df['category'] == 'ART_AND_DESIGN']
```

## 2)list all the  paid app

```python
print('Free : ',len(df[df['type'] == 'Free']))
print('Paid : ',len(df)-len(df[df['type'] == 'Free']))
```

## 3) list number of apps in each category

```python
for name in df['category'].unique():
    print(name,' : ',len(df[df['category']==name]))
```


## 4) Top 5 categories in terms of number of applications

```python
data = []
for name in df['category'].unique():
    data.append([name,len(df[df['category'] == name])])

df_pr = pd.DataFrame(data,columns=['name','freq'])
```

```python
df_pr.sort_values(by ='freq',ascending = False).head(5)
```

## 5) Top 5 categories in terms of number of applications

```python
data = []
for name in df['content_rating'].unique():
    data.append([name,len(df[df['content_rating'] == name])])

df_pr = pd.DataFrame(data,columns=['name','freq'])
df_pr.sort_values(by ='freq',ascending = False).head(5)
```

## 6)list the unrated apps  in content_rating

# 8) average price of apps in each content rating

```python
for name in df['content_rating'].unique():
    print(name , int(df[df['content_rating'] == name]['price'].mean()))
```

# 9) maximum price of apps in each content rating

```python
for name in df['content_rating'].unique():
    print(name , df[df['content_rating'] == name]['price'].max())
```

# 10) Minimum price of apps in each content rating

```python
for name in df['content_rating'].unique():
    print(name , df[df['content_rating'] == name]['price'].min())
```

# 11) Top 5 categories in terms of total number of reviews

```python
data = []
for name in df['category'].unique():
    data.append([name,df[df['category'] == name]['review'].sum()])

df_pr = pd.DataFrame(data,columns = ['category','reviews']).sort_values(by='reviews',ascending =False).head(5)
```

```python
df_pr
```

## 12) Number of apps and number of reviews ratio 

```python
for name in df['content_rating'].unique():
    print(name, int(df[df['content_rating'] == name]['review'].sum() / len(df[df['content_rating'] == name])))
    
```

# 13) Top 5 number of apps and number of reviews ratio in category

```python
data = []
for name in df['content_rating'].unique():
    data.append([name, int(df[df['content_rating'] == name]['review'].sum() / len(df[df['content_rating'] == name]))])
    
df_pr = pd.DataFrame(data,columns = ['category','ratio'])
```

```python
df_pr.head()
```

# 14) list the Paid apps whose category is FAMILY ,content-rating is everyone and rating == 5

```python
df_pr = df[df['category'] == 'FAMILY']
df_pr = df_pr[df_pr['type'] == 'Paid']
df_pr = df_pr[df_pr['content_rating'] == 'Everyone']
df_pr = df_pr[df_pr['rating'] == 5]
```
