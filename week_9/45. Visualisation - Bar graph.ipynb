# Level -9 ( Advanced Stacked Graph)

```python
name = ['Ashish','Raj','Rohan','Rahul']
eng = np.array([46,56,76,78])
his = np.array([88,56,76,88])
mth = np.array([73,63,67,87])
sci = np.array([88,67,87,88])
```

```python
plt.bar(name, sci,color = 'g')#science  - Green
plt.bar(name, mth , bottom = sci,color = 'b')
plt.bar(name, his, bottom = sci + mth,color = 'r')
plt.bar(name, eng, bottom = sci +mth+his,color = 'orange')

plt.title('comparing student marks in different subjects ')

plt.xlabel('Names')
plt.ylabel('Marks')

plt.legend(['Science','Maths','History','English'])

plt.show()
```
