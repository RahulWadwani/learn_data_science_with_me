```python
marks = [65,75,85,95,45,15,25,35,77,88,99,44,55,66,71,72,75,79,76,74]

fig,ax = plt.subplots()
ax.hist(marks,bins = 10)

plt.xlabel(['Marks'])
plt.ylabel(['Frequency'])

plt.title('Marks Distribution')

plt.show()
```

# Level 5 -(Normal Distribution on Google playstore dataset)

```python
df = pd.read_csv('playstore.csv')
df.head()
```

```python
# right skewed data outliers
data = list(df['rating'])

fig, axs = plt.subplots()
axs.hist(data,bins = 30)
plt.show()
```

# Level - 6 (Data Distribution of Ratings)-Scratch

```python
data = list(df['rating'])

ratings = []
for rating in set(data):
    c = 0
    for i in range(len(data)):
        if data[i] == rating:
            c += 1  
    ratings.append([rating,c])
df_rating = pd.DataFrame(ratings, columns = ['rating','freq'])
rating = list(df_rating.sort_values(by='rating')['rating'])
freq = list(df_rating.sort_values(by='rating')['freq'])
```

```python
plt.bar(rating,freq)
plt.show()
```

# Level - 8 (making it to normally distributed (rating >= 3.5 and rating <5)

```python
data =[]
for i in df['rating']:
    if i >= 3.5 and i < 5:
        data.append(i)

fig ,axs = plt.subplots()
axs.hist(data,bins = [3.5,3.6,3.7,3.8,3.9,4.0,4.1,4.2,4.3,4.4,4.5,4.6,4.7,4.8,4.9,5])
plt.show()
```
