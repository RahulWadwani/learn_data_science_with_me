```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Data-Science-Specialization/main/Week%20-12/IMDB%20-%20One%20Hot%20Encoding%20Genre/imdb.csv')
df.head()
```

```python
# 1.1) removing columns 
```

```python
del df['Certificate']
del df['Runtime']
del df['Meta_score']
del df['Director']
del df['Star1']
del df['Star2']
del df['Star3']
del df['Star4']
del df['No_of_Votes']
del df['Gross']
del df['Poster_Link']
del df['Released_Year']
del df['IMDB_Rating']
del df['Overview']
df.head()
```

```python
# 1.2) checking the null values 
```

```python
df.isnull().sum()
```

```python
# 1.3) Finding all the genre 
```

```python
genre = []
for i in df['Genre']:
    genre += (i.split(','))
genre = [i.strip() for i in genre]

genre = (list(set(genre)))
```

# representation 
### this is termed to be as one hot representation 


####                      'Action'       'Crime'         'Drama'
#### 'Action'                [1,              0,                0]
#### 'Crime'                 [0,              1,                0]
#### 'Drama'                [0,              0,                1]
#### 'Drama , Crime '  [0,              1,                1]
#### 'Action, Crime '   [1,              1,                0]

# 1.4 one hot conversion 

```python
one_hot = []
for i in df['Genre']:
    i = [j.strip() for j in i.split(',')]
    enc = [0 for i in range(21)]
        
    for gen in i :
        enc[genre.index(gen)] = 1
    one_hot.append(enc)
```

```python
len(one_hot)
```

```python
# 1.5 one hot conversion - verification 
```

```python
for i in range(len(df['Genre'])):
    x1 = len(df['Genre'][i].split(','))
    x2 = list(one_hot[i]).count(1)
    if x1!=x2:
        print('not Equal')
```

```python
# converting one_hot data to numpy array
```

```python
one_hot = np.array(one_hot)
one_hot.shape
```

```python
# 1.7 adding one_hot data to dataframe
```

```python
for i in range(21):
    df[genre[i]] = one_hot[:,i]
df.head()
```

# prediction 

##  2.1 with exact genre

```python
user = df[df['Series_Title']== 'The Godfather'].values[0][2:]
```

```python
user
```

```python
for i in df.values:
    if (list(user - i[2:]).count(0)) == len(user):
        print(i[0],'|',i[1])
```

```python
### 2.2) with better recommendation with genre
```

```python
def mod(lst):
    s = 0
    for i in lst:
        if i < 0:
            s += -i 
        else:
            s += i
    return s
        
```

```python
user = df[df['Series_Title']== 'The Godfather'].values[0][2:]

rel =[]

for i in df.values:
    rel.append(mod(i[2:]-user))
df['rel'] = rel
df.sort_values(by='rel').head()
```
